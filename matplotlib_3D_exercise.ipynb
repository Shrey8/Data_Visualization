```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

* Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>
* hyperbolic paraboloid equation : z = x*x - y*y

```python
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)

def func(x,y):
    z = x*x - y*y
    return z

Z = func(X,Y)
```

```python
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# a wireframe
ax.plot_wireframe(X, Y, Z, color='r')
plt.show()
```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)

def func(x,y):
    z = x*x + y*y
    return z

Z = func(X,Y)

fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# a wireframe
ax.plot_surface(X, Y, Z, cmap ='jet')
plt.show()

```

* Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>
* Create also contour projection on the plane z=0 (HINT: use parameter 'zdir' of contour plot)
* eliptical paraboloid : z = x*x + y*y

```python
# contour plot

# surface plot
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)

def func(x,y):
    z = x*x + y*y
    return z

Z = func(X,Y)

fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# a wireframe
ax.contour(X, Y, Z,zdir= 'Z')
plt.show()


```
