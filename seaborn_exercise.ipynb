```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

```python
df.head()
```

# Objective: Plot regression line between attack and defense

```python
sns.lmplot(x= 'Defense', y= 'Attack' , data = df)
plt.show()
```

# Objective: Plot regression line between attack and defense with respect to generation

```python
sns.lmplot(x= 'Defense', y= 'Attack' , data = df , col = 'Generation')
plt.show()
```

# Objective: Make boxplot for each feature of dataframe (without Generation and Legendary features) in one plot

```python
df.head()
```

```python
plt.subplot(3,3,1)
sns.boxplot(x = 'Total', data = df)

plt.subplot(3,3,2)
sns.boxplot(x = 'HP', data = df)

plt.subplot(3,3,3)
sns.boxplot(x = 'Attack', data = df)

plt.subplot(3,3,4)
sns.boxplot(x = 'Defense', data = df)

plt.subplot(3,3,5)
sns.boxplot(x = 'Sp. Atk', data = df)

plt.subplot(3,3,6)
sns.boxplot(x = 'Sp. Def', data = df)

plt.tight_layout()
```

# Objective: Plot correlation heatmap of pokemon dataframe (not include features Generation and Legendary)

```python
# Heatmap
# compute correlations between features
df_corr = df.iloc[:,:-2].corr()

# plot the correlations
sns.heatmap(df_corr)
plt.title('Correlation plot')
plt.show()
```
